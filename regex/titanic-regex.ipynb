```python
def search_na(pattern, x):
    try:
        text = re.search(pattern, x).group()
    except:
        text = np.nan
    return text
```

```python
titanic
```

```python
titanic['home.dest'].apply(lambda x: search_na(pattern, x))
```

```python
titanic['home.state'] = titanic['home.dest'].apply(lambda x: search_na(pattern, x))
```
